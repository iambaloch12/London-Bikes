```python
#change column names that i want to use using dictionary
dic = {
    'timestamp':'time',
    'cnt':'count',
    't1':'real_time',
    't2':'temp_feel',
    'hum':'humidity',
    'wind_speed':'wind_speed_kph',
    'weather_code':'weather',
    'is_holiday':'is_holiday',
    'is_weekend':'is_weekend',
    'season':'season'
}

#renaming the columns to the specified columns
bikes.rename(dic,axis=1,inplace=True)
```

```python
bikes
```

```python
#change the himidity vales to the percentage
bikes.humidity = bikes.humidity/100
```

```python
#creating a season dictionary so that we can map the actual written values
season_dict = {
    '0.0':'Spring',
    '1.0':'Summer',
    '2.0':'Autumn',
    '3.0':'Winter'
}

#creating a weather dictionary so that we can map the actual written values
weather_dict = {
    '1.0':'Clear',
    '2.0':'Scattered Clouds',
    '3.0':'Broken Clouds',
    '4.0':'Cloudy',
    '7.0':'Rain',
    '10.0':'Rain With Thunderstorm',
    '26.0':'Snowfall'
}

#changing the season column type to string
bikes.season = bikes.season.astype('str')
#mapping the values
bikes.season = bikes.season.map(season_dict)

#changing the weather column type to string
bikes.weather = bikes.weather.astype('str')
#mapping the values
bikes.weather = bikes.weather.map(weather_dict)
```
